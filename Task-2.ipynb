```python
df = df[['coaster_name','Location', 'Status', 'Manufacturer', 'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',  'speed_mph','height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df = df.rename(columns = {'coaster_name':'Coaster_Name',
                    'year_introduced':'Year_Introduced',
                    'latitude':'Latitude',
                     'longitude':'Longitude',
                     'opening_date_clean':'Opening_Date',
                     'speed_mph':'Speed_mph',
                     'height_ft':'Height_ft',
                    'Gforce_clean':'Gforce'})
```

```python
df = df.loc[~df.duplicated(subset = ['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True)
```

# Feature Understanding

```python
df['Year_Introduced'].value_counts()
```

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(kind = 'bar', title = 'Top Years Coasters Introduced')
ax.set_xlabel('Year_Introduced')
ax.set_ylabel('Counts')
```

```python
ax = df['Speed_mph'].plot(kind = 'hist', bins = 20, title = 'Coaster Speed(mph)')
ax.set_xlabel('Speed')
```

```python
ax = df['Speed_mph'].plot(kind = 'kde',  title = 'Coaster Speed(mph)')
ax.set_xlabel('Speed')
```

## Feature Relationships

```python
df.plot(kind = 'scatter',
       x = 'Speed_mph',
       y = 'Height_ft',
        title = 'Coaster Speed vs. Height'
       )
```

```python
sns.scatterplot( x = 'Speed_mph',y = 'Height_ft',data = df, hue = 'Year_Introduced')
```

```python
sns.pairplot(df, vars= ['Year_Introduced', 'Speed_mph','Height_ft',
                        'Inversions_clean','Gforce'],hue = 'Type_Main')
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph','Height_ft',
                        'Inversions_clean','Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr,annot = True)
```
